```python
import pandas as pd
Crime = pd.read_excel("Crime.xlsx")
Crime
```

```python
Crime = Crime.drop('Unnamed: 0', axis =1)
```

```python
Crime
```

```python
cols=['Age', 'S', 'Ed', 'Ex0', 'Ex1', 'LF', 'M', 'N', 'NW', 'U1', 'U2',
       'W', 'X']
```

```python
from sklearn.preprocessing import StandardScaler
features = cols
# Separating out the features
x = Crime.loc[:, features].values
# Separating out the Crime Rate
y = Crime.loc[:,['Crime Rate']].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, Crime[['Crime Rate']]], axis = 1)
```

```python
finalDf 
```

```python
print(Crime['Crime Rate'].min(), Crime['Crime Rate'].max())
```

```python
p1_low=finalDf[finalDf['Crime Rate'].between(34.2,55)]['principal component 1']
p2_low=finalDf[finalDf['Crime Rate'].between(34.2,55)]['principal component 2']
p1_average=finalDf[finalDf['Crime Rate'].between(55,89.2)]['principal component 1']
p2_average=finalDf[finalDf['Crime Rate'].between(55,89.2)]['principal component 2']
p1_high=finalDf[finalDf['Crime Rate'].between(89.2,144)]['principal component 1']
p2_high=finalDf[finalDf['Crime Rate'].between(89.2,144)]['principal component 2']
```

```python
plt.scatter(p1_low,p2_low,color = 'blue')
plt.scatter(p1_average,p2_average,color = 'green')
plt.scatter(p1_high,p2_high,color = 'red')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Admission = pd.read_excel('Admission.xlsx')
Admission = Admission.drop('Unnamed: 0', axis =1)
Admission.head()
```

```python
x = Admission.drop('Recu',axis=1)
y = Admission['Recu']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# importing module
from sklearn.linear_model import LinearRegression
# creating an object of LinearRegression class
LR = LinearRegression()
# fitting the training data
LR.fit(x_train,y_train)
```

```python
y_prediction =  LR.predict(x_test)
y_prediction
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_prediction, y_test)
```
